```python
import json
```

READING RECORD.JSON FILE AND STORING IT IN LOCAL VARIABLE

```python
fd=open("record.json",'r') 
r=fd.read() 
fd.close
data=json.loads(r)
```

PRINTING ITEMS AVAILABLE IN INVENTORY

```python
print("Welcome") 
print("We have these items available:"+"\n") 
for i in data:
   print("Product_id:"+i+"-",data[i])
```

PURCHASING ITEM

```python
 
Prod_id=input("Enter Product Id: ") 
quan=int(input("Enter No. of Quantity u want to purchase: "))
 
if Prod_id not in data :
   print("Sorry, We don't have any item of this product id.")
   Prod_id="0"
   quan=0
elif quan>data[Prod_id]["Quantity"]:
   print("Sorry we have only "+str(data[Prod_id]["Quantity"]) +" units left")
   Prod_id="0"
   quan=0
else:
   print("***************************************") 
   print("Product_Id: "+Prod_id) 
   print("Product Name: "+data[Prod_id]["Name"]) 
   print("Price: ", data[Prod_id]["Price"]) 
   print("Unit purchased: ", quan) 
   print("Total amount: ", quan*data[Prod_id]["Price"]) 
   print("discount: ", .1*quan*data[Prod_id]["Price"]) 
   print("Bill Amount", ((quan*data[Prod_id]["Price"]) -(.1*quan*data[Prod_id]["Price"]))) 
   print("****************************************")
print("Thanks for Shopping") 
print("Have a Nice day")
```

UPDATING INVENTORY

```python
data[Prod_id]["Quantity"]=((data[Prod_id]["Quantity"]) -quan)
js=json.dumps(data) 
fd=open("record.json", 'w') 
fd.write(js) 
fd. close()
```

CREATING SALES DATA

```python
fd=open("sales.json", 'r') 
txt=fd.read()  
fd.close
Sale=json.loads(txt)
```

```python
import time
Time=time.ctime() 
 
Sale[str(len(Sale)+1)]={'Product Name':data[Prod_id]["Name"],"Product id":Prod_id, "Quantity":quan, "Amount":((quan*data[Prod_id]["Price"]) -(.1*quan*data[Prod_id]["Price"])),"Timing":Time}
js=json.dumps(Sale) 
fd=open("sales.json", 'w') 
fd.write(js) 
fd.close()
```

SHOWING SALES DATA

```python
x=open("sales.json", 'r') 
a=x.read() 
x.close() 
sales_data=json.loads(a) 
for i in sales_data:
   print("Transaction Id:"+i+"-", sales_data[i])
```
